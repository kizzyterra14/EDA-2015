Implementação do Quicksort

```python
def partition(L):
    
    pivot = L[0]
    lower = 0
    for current in range (1,len(L)):
        if L[current] < pivot:
            lower +=1
            temp =  L[lower]
            L[lower] = L[current]
            L[current] = temp
            
    L[0] = L[lower]
    L[lower] = pivot
    
    #Este for foi incluído para adaptar o algoritmo para entradas com números duplicados, 
    #a complexidade aumenta de k*n para c*2n mas ainda é O(n)
    for current in range (lower+1,len(L)):
        if L[current] == pivot:
            lower +=1
            temp =  L[lower]
            L[lower] = L[current]
            L[current] = temp
    
    return L, lower
        
```

```python
partition([3,3,2,8,5,1,4,7,6])
```

```python
def quicksort(L, filename):
    if len(L) < 2:
        return L
    else:
        (lst, pivot_index) = partition(L)
        fl=open(filename, 'a')
        fl.write(str(float(pivot_index+1)/float(len(L)))+'\n')
        return quicksort(lst[:pivot_index],filename) + [L[pivot_index]] + quicksort(lst[pivot_index+1:],filename)
        
```

```python
quicksort([3,3,2,8,5,1,4,7,6],'teste.txt')
```

```python
quicksort([],'teste.txt')
```

```python
quicksort([3],'teste.txt')
```

```python
quicksort([3,6],'teste.txt')
```

Analisando escolha dos pivôs:

Gerando arquivo com 'len_file' listas com 'len_list' numeros aleatorios cada

```python
import random, tempfile 

def modfile(len_file, len_list):
    with tempfile.NamedTemporaryFile(mode = 'w+', delete = False) as newFile:
        for i in range(0,len_file*len_list):
            newFile.write(str(random.randint(0, 1000))+'\n')
    return newFile.name
```

```python
len_file = 3000 #Número N de rodadas
len_list = 1000 #Número M de números aleatórios
tempfile = modfile(len_file, len_list)
```

A seguir iremos recuperar as listas do arquivo e fazer uma chamada do quicksort para cada uma:

```python
f = open(tempfile)
lines = f.readlines()
```

```python
#Recuperando listas do arquivo
lists = []
line = 0
list = 0
while line < len(lines):
    L=[]
    for i in range(0,len_list):
        L.insert(i,int(lines[line]))
        i+=1
        line += 1
    lists.insert(list, L)
    list+=1
```

```python
filename = 'pivots.txt'
for lst in lists:
        quicksort(lst, filename)
```

A função quicksort armazena no arquivo 'filename' a escolha de pivos das chamadas recursivas, a partir da informação salva no arquivo podemos analisar a media de boas escolhas dos pivos no quicksort

```python
#Recuperando pivos do arquivo
f = open(filename)
pivots_lines = f.readlines()
pivots = []
line = 0
while line < len(pivots_lines):
    pivots.insert(line,float(pivots_lines[line]))
    line += 1
```

```python
import pandas as pd
df= pd.DataFrame(pivots)
df.columns = ['pivos']

def analysis(pivot):
    if pivot < 0.75 and pivot > 0.25:
        return 'boa'
    else:
        return 'ruim'
    
df['escolha'] = df['pivos'].apply(analysis)
df
```

```python
#Analisando porcentagem de boas escolhas
count = 0
for ratio in df['pivos']:
    if ratio < 0.75 and ratio > 0.25:
        count+=1

print float(count)/float(len(pivots))
    
```

O resultado obtido foi de aproximadamente 48% de boas escolhas. Esta porcentagem é suficiente razoável para garantir O(n log n) no Quicksort
